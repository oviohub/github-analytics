```python
def filter_weeks(contribution_weeks):
    return [week for week in contribution_weeks if (week['a'] + week['d'] + week['c']) > 0]

def get_stats_by_user(g, repo):
    repo = g.get_repo(repo)
    stats = repo.get_stats_contributors()
    return {
        stat.author.login: filter_weeks(stat.raw_data['weeks'])
        for stat in stats
    }
```

```python
stats_by_user = get_stats_by_user(g, 'wfp/vampire-backend')
```

```python
def add_weeks(stats, weeks):
    counter = collections.Counter()
    for stat in stats:
        if weeks == None or stat['w'] in weeks:
            counter.update({'a': stat['a'], 'd': stat['d'], 'c': stat['c']})
    return counter


def stats_by_user_by_week(stats_by_user, weeks):
#     for user, stats in stats_by_user.items():
#         print(user, add_weeks(stats, weeks))
    return {user: add_weeks(stats, weeks) for user, stats in stats_by_user.items()}
```

```python
stats_by_user_by_week(stats_by_user, weeks=[1544313600, 1544918400, 1545523200, 1546128000])
```

```python
total_by_user = {}
for project in projects:
    print(project)
    stats_by_user = get_stats_by_user(g, project.replace('https://github.com/', ''))
    stats_by_user_by_week_data = stats_by_user_by_week(stats_by_user, weeks=None)
    for user in stats_by_user_by_week_data:
        if user in total_by_user:
            total_by_user[user].update(stats_by_user_by_week_data[user])
        else:
            total_by_user[user] = stats_by_user_by_week_data[user]
```
